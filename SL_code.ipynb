```python
from scipy.interpolate import interp1d
```

```python
def f_n(x,X,Y):
    if x in X: 
        return Y[X == x]
    denoms = 1/abs(X - x*np.ones(len(X))) # 1/|xi - x|
    if sum(denoms) == 0: 
        return 0
    else:  
        return sum(denoms*Y)/sum(denoms)
    
def HK_plot(x,f,xnew):
    y = f(x)
    f_n_values = []
    for i in range(len(xnew)):
        f_n_values.append(f_n(xnew[i], x,y))
    y_interp = interp1d(x, y)
    plt.plot(x, y, 'o', xnew, f_n_values, "--", xnew, f(xnew), "-")
#     plt.plot(x, y, 'o', xnew, f_n_values, "--", xnew, y_interp(xnew), "--", xnew, f(xnew), "-")
    plt.legend(['data', 'Hilbert estimate', "original function"], loc='best')
    plt.title(f"Hilbert estimate, n = {len(x)}. ")
```

```python
x = np.linspace(0, 1, num=3, endpoint=True)
xnew = np.linspace(0, 1, num=1000, endpoint=True)
f = np.sin
HK_plot(x,f,xnew)
```
